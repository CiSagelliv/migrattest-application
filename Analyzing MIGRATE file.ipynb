```python
import re
import os
from itertools import islice
from itertools import permutations
from itertools import combinations
import pandas as pd
```

```python
#Read file and pass its lines to list
with open('outfile-cortado.txt','r') as origin_file:
    #Create an empty list
    file_rows = []
    for rows in origin_file:
        file_rows.append(rows)
    print(file_rows)
```

```python
#Find words -> function
def find_groups(substr, infile):
    with open(infile) as origin:
        for line in origin:
            if substr in line:
                print(line)
                
find_groups('Population','outfile-cortado.txt')
find_groups('Total of','outfile-cortado.txt')
```

```python
#find percentiles -> function
def find_percentiles(substr, infile):
    with open(infile) as origin:
        for line in origin:
            if substr in line:
                print(line)

find_percentiles('Lower percentiles','outfile-cortado.txt')
find_percentiles('Upper percentiles','outfile-cortado.txt')
```

```python
#find indexes in file_rows list
start_index_group = file_rows.index('Population                   Locus   Gene copies    \n')
end_index_group = file_rows.index('    Total of all populations     1    566 (0)\n')
index_lower = file_rows.index('Parameter                          Lower percentiles\n')
index_upper = file_rows.index('Parameter                          Upper percentiles\n')
print(start_index_group, end_index_group, index_lower, index_upper)
```

```python
#This cell must be executed once
#Group csv file
groups_filename = 'groups'
index_1 = 0

while os.path.exists(f"{groups_filename}{index_1}.csv"):
    index_1 += 1

with open('outfile-cortado.txt', 'r') as origin_file:
    for line in islice(origin_file, start_index_group + 4, end_index_group):
        with open(f"{groups_filename}{index_1}.csv", 'a') as group_file:
            print(re.sub("\s+",",", line.strip()))
            group_file.write(re.sub("\s+",",", line.strip()) + '\n')
```

```python
#Groups dataframe with unnecessary data
groups_df = pd.read_csv(f"{groups_filename}{index_1}.csv", header=None)
groups_df
```

```python
#This cell must be executed once
#Remove NaN values
groups_df = groups_df.dropna()
#Remove 'Locus' column
groups_df = groups_df.drop(2,1)
#Remove 'missing (0)' column
groups_df = groups_df.drop(4,1)
groups_df
```

```python
#Get total of groups and its n-value
#Change data types str -> int
total_per_group = list(map(int, (groups_df[3])))
print(total_per_group, len(total_per_group))
```

```python
#Start and end lower percentiles
range_lower_start = index_lower + 4 + len(total_per_group)
range_lower_end = range_lower_start + len(list(permutations(range(len(total_per_group)),2)))
print(range_lower_start, range_lower_end)
```

```python
#Start and end upper percentiles
range_upper_start = index_upper + 4 + len(total_per_group)
range_upper_end = range_upper_start + len(list(permutations(range(len(total_per_group)),2)))
print(range_upper_start, range_upper_end)
```

```python
#This cell must be executed once
#lower percentiles csv
lower_percentiles_filename = 'lower_percentiles'
index_2 = 0

while os.path.exists(f"{lower_percentiles_filename}{index_2}.csv"):
    index_2 += 1

with open('outfile-cortado.txt', 'r') as origin_file:
    for line in islice(origin_file, range_lower_start, range_lower_end):
        with open(f"{lower_percentiles_filename}{index_1}.csv",'a') as lower_file:
            print(re.sub("\s+", ",", line.strip()))
            lower_file.write(re.sub("\s+", ",", line.strip()) + '\n')
```

```python
#This cell must be executed once
#upper percentiles csv
upper_percentiles_filename = 'upper_percentiles'
index_3 = 0

while os.path.exists(f"{upper_percentiles_filename}{index_3}.csv"):
    index_3 += 1

with open('outfile-cortado.txt', 'r') as origin_file:
    for line in islice(origin_file, range_upper_start, range_upper_end):
        with open(f"{upper_percentiles_filename}{index_3}.csv", 'a') as upper_file:
            print(re.sub("\s+", ",", line.strip()))
            upper_file.write(re.sub("\s+", ",", line.strip()) + '\n')
```

```python
#lower percentiles dataframe
lower_df = pd.read_csv(f"{lower_percentiles_filename}{index_1}.csv",header=None)
lower_df.columns = ['Pairs','P1','P2','P3','P4','MLE']
lower_df
```

```python
#upper percentiles dataframe
upper_df = pd.read_csv(f"{upper_percentiles_filename}{index_3}.csv",header=None)
upper_df.columns = ['Pairs','MLE','P5','P6','P7','P8']
upper_df
```

```python
#This code must be executed once
#Delete repeated columns
upper_df = upper_df.drop(['MLE'], axis=1)
upper_df = upper_df.drop(['Pairs'], axis=1)
upper_df
```

```python
#percentiles dataframe
percentiles_df = pd.concat([lower_df, upper_df],axis=1)
percentiles_df
```

```python
#Delete * from rows
percentiles_df['P1'] = percentiles_df['P1'].str.replace('*','')
percentiles_df['P2'] = percentiles_df['P2'].str.replace('*','')
percentiles_df['P3'] = percentiles_df['P3'].str.replace('*','')
percentiles_df['P4'] = percentiles_df['P4'].str.replace('*','')
percentiles_df['P5'] = percentiles_df['P5'].str.replace('*','')
percentiles_df['P6'] = percentiles_df['P6'].str.replace('*','')
percentiles_df['P7'] = percentiles_df['P7'].str.replace('*','')
percentiles_df['P8'] = percentiles_df['P8'].str.replace('*','')
percentiles_df
```

```python
#Change data types object -> numeric
percentiles_df['P1'] = pd.to_numeric(percentiles_df['P1'], errors = 'coerce')
percentiles_df['P2'] = pd.to_numeric(percentiles_df['P2'], errors = 'coerce')
percentiles_df['P3'] = pd.to_numeric(percentiles_df['P3'], errors = 'coerce')
percentiles_df['P4'] = pd.to_numeric(percentiles_df['P4'], errors = 'coerce')
percentiles_df['P5'] = pd.to_numeric(percentiles_df['P5'], errors = 'coerce')
percentiles_df['P6'] = pd.to_numeric(percentiles_df['P6'], errors = 'coerce')
percentiles_df['P7'] = pd.to_numeric(percentiles_df['P7'], errors = 'coerce')
percentiles_df['P8'] = pd.to_numeric(percentiles_df['P8'], errors = 'coerce')
percentiles_df['MLE'] = pd.to_numeric(percentiles_df['MLE'], errors = 'coerce')
percentiles_df
```

```python
#Get the sum of significant percentiles rows
mixed_percentiles = percentiles_df[['Pairs','P3','MLE','P6']]
mixed_percentiles['Sum_of_values_1'] = mixed_percentiles.sum(axis=1)
mixed_percentiles
```

```python
mixed_percentiles['Sum_of_values_2'] = mixed_percentiles['Sum_of_values_1']
mixed_percentiles
```

```python
#Get pair's first element (total)
first_element = list(combinations((total_per_group),len(total_per_group)-1))
print(first_element)
```

```python
#Reverse the list 
first_element.reverse()
print(first_element)
```

```python
#Create list of tuples, first_elements returns a list of tuples
first_element_list = []

for tuple_element in first_element:
    for list_element in tuple_element:
        first_element_list.append(list_element)
        
print(first_element_list, len(first_element_list))
```

```python
mixed_percentiles['First_total'] = first_element_list
mixed_percentiles
```

```python
second_element_list = [element for element in total_per_group for i in range(len(total_per_group) -1)]
print(second_element_list)
```

```python
mixed_percentiles['Second_total'] = second_element_list
mixed_percentiles
```
